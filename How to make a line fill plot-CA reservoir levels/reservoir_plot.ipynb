```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read data from csv
data_avg = pd.read_csv('average_reservoir.csv')
data_now = pd.read_csv('current_reservoir.csv')
data_avg.head()
```

```python
# Transfer data into numpy arrays
y_avg_values = data_avg.Level.to_numpy()
y_now_values = data_now.Level.to_numpy()

x_avg = pd.to_datetime(data_avg.Date)
x_now = pd.to_datetime(data_now.Date)

x_avg_values = x_avg.to_numpy()
x_now_values = x_now.to_numpy()
```

```python
# Set figure size
plt.rcParams['figure.figsize'] = (4,8)
plt.rcParams['figure.dpi'] = 150

fig, ax = plt.subplots()
ax.fill_between(x_avg_values, y_avg_values, alpha=0.2)
ax.plot(x_now_values, y_now_values, color='k')
ax.set_ylabel('Fractional Capacity')

# Choose which months to label on x-axis
import matplotlib.dates as mdates
ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=1))

# Remove the y-axis and spines
ax.yaxis.set_visible(False)
ax.spines[['left','right','top']].set_visible(False)

# Format the dates
# %m is month, %d is day, %Y is 4 digit year, 
# %y is 2 digit year, %b is the abbreviated month

from matplotlib.dates import DateFormatter
date_form = DateFormatter('%b.\n%Y')
ax.xaxis.set_major_formatter(date_form)

# The hard way to add text to plots
#ax.text(x_avg_values[8],0.42,'Reservoir Levels')
plt.savefig('reservoir_levels.png')

plt.show()
```
